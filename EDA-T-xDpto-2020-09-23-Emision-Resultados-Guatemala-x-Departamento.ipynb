```python
df_dep = data.drop(columns = ['codigo_departamento', 'municipio', 'codigo_municipio', 'poblacion'],
                  axis = 1)
df_dep['departamento'] = df_dep['departamento'].str.title()
df_dep.head()
```

```python
# tenemos 198 filas pero 24 departamentos diferentes
# se procede a combinar filas y obtener la sumatoria
# Pivot: se realiza el traslado de columnas por filas para obtener la sumatoria por dep.
total_columnas = df_dep.columns[1:].to_numpy()
#total_filas = df_dep['departamento'][total_filas].keys().to_numpy()
#sumarize rows
df = df_dep.groupby('departamento').sum()
#df_dep_p = pd.DataFrame(df_dep[total_columnas].values.reshape(1,-1), 
 #                   index=['Market 1 Order'], 
  #                  columns=total_filas)
df.head()
```

```python
final_df = df.transpose()
final_df.rename(columns={'departamento':'date'},inplace= True)
```

```python
final_df.drop("Sin Datos", axis=1, inplace=True)
final_df.drop("Sin Dato", axis=1, inplace=True)
final_df.drop("Otros", axis=1, inplace=True)
```

```python
# Crear columna Total
final_df['Total'] = final_df.sum(axis = 1)
```
